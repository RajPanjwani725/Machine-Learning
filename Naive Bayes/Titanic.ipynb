```python
import pandas as pd
```

```python
df=pd.read_csv("train.csv")
```

```python
df
```

```python
data=df[["Pclass","Sex","Age",'Fare','Survived']]
```

```python
data
```

```python
target=data.Survived
target
```

```python
features=data.drop('Survived',axis=1)
features
```

### One hot encoding

```python
dummies=pd.get_dummies(features.Sex)
```

```python
dummies
```

```python
features=pd.concat([features,dummies],axis=1)
```

```python
features=features.drop('Sex',axis=1)
features
```

```python
features.isnull().sum()
```

# train test split

```python
from sklearn.model_selection import  train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(features,target,test_size=0.3,random_state=1)
```

```python
X_test
```

# Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model=GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

```python
y_predict=model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_predict,normalize=True)
```

```python
# Correct = (Add diagonal) 138+80=218

#Total=138+80+28+21

# %correct= (218/267)*100 = 81.64%

confusion_matrix(y_test,y_predict)
```
